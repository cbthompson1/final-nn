```python
# Imports

import matplotlib.pyplot as plt
from nn import nn
import numpy as np
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split

seed = 1000
```

```python
# Load the digits dataset through sklearn and read in data and generate training and validation sets.

digits = load_digits()
X = digits.data # (1797, 64)
y = digits.target # (1797,)

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=seed)
```

```python
# Generate an instance of your `NeuralNetwork` class with a 64x16x64 autoencoder architecture and train your autoencoder on the training data.

epochs = 10000

neural_network = nn.NeuralNetwork(
    [{'input_dim': 64, 'output_dim': 16, 'activation': 'relu'}, {'input_dim': 16, 'output_dim': 64, 'activation': 'sigmoid'}],
    lr = 0.01,
    seed=seed,
    batch_size=50,
    epochs=epochs,
    loss_function="mean_squared_error"
)

train_loss, val_loss = neural_network.fit(X_train, y_train, X_val, y_val)
```

```python
# Plot your training and validation loss by epoch.

plt.figure(figsize=(10, 6))
plt.plot(train_loss, 'b-', label='Training Loss')
plt.plot(val_loss, 'r-', label='Validation Loss')
plt.title("Train and Validation Loss Curves")
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# MORE LAYERS!

epochs = 10000

neural_network = nn.NeuralNetwork(
    [
        {'input_dim': 64, 'output_dim': 32, 'activation': 'relu'},
        {'input_dim': 32, 'output_dim': 16, 'activation': 'relu'},
        {'input_dim': 16, 'output_dim': 32, 'activation': 'relu'},
        {'input_dim': 32, 'output_dim': 64, 'activation': 'relu'}
    ],
    lr = 0.05,
    seed=seed,
    batch_size=100,
    epochs=epochs,
    loss_function="mean_squared_error"
)

train_loss, val_loss = neural_network.fit(X_train, y_train, X_val, y_val)
```

```python
# Plot your training and validation loss by epoch.

plt.figure(figsize=(10, 6))
plt.plot(train_loss, 'b-', label='Training Loss')
plt.plot(val_loss, 'r-', label='Validation Loss')
plt.title("Train and Validation Loss Curves")
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# What if we add more layers and use sigmoid?

epochs = 10000

neural_network = nn.NeuralNetwork(
    [
        {'input_dim': 64, 'output_dim': 48, 'activation': 'relu'},
        {'input_dim': 48, 'output_dim': 32, 'activation': 'sigmoid'},
        {'input_dim': 32, 'output_dim': 16, 'activation': 'relu'},
        {'input_dim': 16, 'output_dim': 32, 'activation': 'relu'},
        {'input_dim': 32, 'output_dim': 48, 'activation': 'sigmoid'},
        {'input_dim': 48, 'output_dim': 64, 'activation': 'relu'}
    ],
    lr = 0.16,
    seed=seed,
    batch_size=1024,
    epochs=epochs,
    loss_function="mean_squared_error"
)

train_loss, val_loss = neural_network.fit(X_train, y_train, X_val, y_val)

plt.figure(figsize=(10, 6))
plt.plot(train_loss, 'b-', label='Training Loss')
plt.plot(val_loss, 'r-', label='Validation Loss')
plt.title("Train and Validation Loss Curves")
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```

```python
print("Model Results:")
print(f"Best Reconstruction Error: {val_loss[-1]:.8f}")
```
